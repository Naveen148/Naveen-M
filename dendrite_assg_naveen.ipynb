```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
col= pd.DataFrame(ohe.fit_transform(data[['species']]).toarray(),columns = data['species'].unique())
data= pd.concat([data,col], axis=1)
data.drop('species',axis=1, inplace=True)
```

```python
feature_dic= df.loc['feature_handling','design_state_data']
feature_dic
```

```python
def feature_handling(feature_handling, column_names,data):
    for col in column_names:
        try:
            if feature_handling[col]['feature_details']['impute_with'] == 'custom':
                data[col] = data[col].fillna(feature_handling[col]['feature_details']['impute_value'])
            elif feature_handling[col]['feature_details']['impute_with'] == 'Average of values': 
                data[col] = data[col].fillna(df[col].mean())
        except KeyError:
            print(col,"feature does not have 'impute_with' provided")
    return df
```
